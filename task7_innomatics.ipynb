```python

#1. Problem 1: Smart Light Controller

# A class to manage a house light system
class HomeGlow:
    def __init__(self, room_label, is_glowing):
        # Unique attributes for the room name and light state
        self.area_name = room_label
        self.light_state = is_glowing

    def show_bulb_info(self):
        # Checking if the light is active or not
        if self.light_state == "ON":
            print(f"{self.area_name} is ON")
        else:
            print(f"{self.area_name} is OFF")

# Setting the inputs
my_room = "Bedroom Light"
action_taken = "ON"

# Creating the object and running the function
smart_bulb = HomeGlow(my_room, action_taken)
smart_bulb.show_bulb_info()
```

```python

#2. Problem 2: Employee ID Card System

# A class to store staff information for an ID badge
class StaffIdentity:
    def __init__(self, full_name, unique_code, office_dept):
        # Setting up the details
        self.staff_member = full_name
        self.identification_no = unique_code
        self.work_section = office_dept

    def print_id_badge(self):
        # Printing the details in a list format
        print("Employee ID Card")
        print(f"Name: {self.staff_member}")
        print(f"ID: {self.identification_no}")
        print(f"Department: {self.work_section}")

# User details [cite: 40, 41, 42]
person_name = "Rahul"
id_tag = "EMP102"
dept_name = "IT"

# Creating the ID card
new_badge = StaffIdentity(person_name, id_tag, dept_name)
new_badge.print_id_badge()
```

```python

#3. Problem 3: Mobile Contact Record

# Class to hold phonebook information
class ContactData:
    def __init__(self, contact_alias, digit_string):
        # Saving the friend's name and their number
        self.saved_name = contact_alias
        self.phone_val = digit_string

    def display_record(self):
        # Outputting the saved details
        print("Contact Saved")
        print(f"Name: {self.saved_name}")
        print(f"Phone: {self.phone_val}")

# Data for the contact
friend_name = "Anita"
cell_no = "9876543210"

# Running the process
new_entry = ContactData(friend_name, cell_no)
new_entry.display_record()
```

```python

#4. Problem 4: Product Price Tag Generator

# Class to generate a store price sticker
class PriceLabel:
    def __init__(self, gadget_title, sticker_price):
        # Assigning values to the object
        self.item_name = gadget_title
        self.cost_val = sticker_price

    def generate_tag(self):
        # Printing with the Rupee symbol
        print(f"Product: {self.item_name}")
        print(f"Price: ₹{self.cost_val}")

# Product details
item_title = "Headphones"
price_tag = 2499

# Outputting the result
store_item = PriceLabel(item_title, price_tag)
store_item.generate_tag()
```

```python
#5. Problem 5: Movie Rating Display System

# Class for a movie database entry
class FilmReview:
    def __init__(self, film_title, star_points):
        # Setting up the movie data
        self.movie_name = film_title
        self.rating_score = star_points

    def show_movie_card(self):
        # Displaying info as requested
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating_score}/5")

# Movie input
target_movie = "Inception"
score_given = 4.8

# Showing the review
review_obj = FilmReview(target_movie, score_given)
review_obj.show_movie_card()
```

```python
#6. Problem 6: Delivery Address Manager

# Class for tracking parcel destinations
class PackageShipment:
    def __init__(self, buyer_name, shipping_spot):
        # Storing who gets it and where
        self.customer = buyer_name
        self.location = shipping_spot

    def show_shipping_info(self):
        # Final output for the logistics system
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.location}")

# Order details
person = "Suman"
city = "Hyderabad"

# Processing the delivery
shipment = PackageShipment(person, city)
shipment.show_shipping_info()
```
