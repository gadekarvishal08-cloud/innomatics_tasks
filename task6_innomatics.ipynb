***Problem 1: Smart Parking Lot Management System***

```python
def check_my_parking(max_spots, car_activity):
    # 'total_right_now' will keep track of cars currently inside
    total_right_now = 0

    for action in car_activity:
        if action == "IN":
            total_right_now = total_right_now + 1
        elif action == "OUT":
            total_right_now = total_right_now - 1

    # Figure out if we have room left
    if total_right_now >= max_spots:
        my_status = "Full / Over Capacity"
    else:
        my_status = "Available"

    print(f"Currently Parked Vehicles: {total_right_now}")
    print(f"Parking Status: {my_status}")

# Testing with sample data
check_my_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

***Problem 2: Online Food Delivery Time Estimator***

```python

def calculate_food_arrival(km_away, road_traffic, sky_condition):
    # Start with a basic estimate: 5 minutes per kilometer
    base_wait = km_away * 5

    # Add extra time for traffic
    if road_traffic == "High":
        base_wait = base_wait + 10
    elif road_traffic == "Medium":
        base_wait = base_wait + 5

    # Add extra time for weather
    if sky_condition == "Rainy":
        base_wait = base_wait + 5

    print(f"Estimated Delivery Time: {base_wait} minutes")

# Testing with sample data
calculate_food_arrival(8, "High", "Rainy")
```

***Problem 3: Movie Theatre Seat Occupancy Analyzer***

```python

def movie_house_stats(total_chairs, ticket_list):
    # Count how many items are in the booked list
    how_many_booked = len(ticket_list)

    # Simple math for percentage
    percent_filled = (how_many_booked / total_chairs) * 100

    # Deciding the status based on percentage
    if percent_filled >= 90:
        the_message = "Housefull"
    elif percent_filled >= 70:
        the_message = "Almost Full"
    else:
        the_message = "Plenty of Seats"

    print(f"Occupancy: {percent_filled}%")
    print(f"Show Status: {the_message}")

# Testing with sample data
# Creating a list with 150 entries to match the prompt
my_booked_seats = [1] * 150
movie_house_stats(200, my_booked_seats)
```

***Problem 4: Cloud Server Load Classification System***

```python
def monitor_my_server(cpu_list):
    # Add up all the numbers and divide by the count
    sum_of_loads = sum(cpu_list)
    number_of_logs = len(cpu_list)
    the_average = sum_of_loads / number_of_logs

    # Check the rules for status
    if the_average >= 80:
        health = "Critical"
    elif the_average >= 50:
        health = "Warning"
    else:
        health = "Normal"

    print(f"Average CPU Load: {the_average}%")
    print(f"Server Status: {health}")

# Testing with sample data
monitor_my_server([45, 60, 70, 85, 90])
```

***Problem 5: Smart Classroom Resource Usage Monitor***

```python

def check_classroom_power(usage_box):
    # We will store names of things used more than 7 hours
    too_much_list = []

    for tool, hours in usage_box.items():
        if hours > 7:
            too_much_list.append(tool)

    # If the list isn't empty, we need an alert
    if len(too_much_list) > 0:
        alert_needed = "Yes"
    else:
        alert_needed = "No"

    # Joining list items into a string for easy reading
    print(f"Overused Resources: {', '.join(too_much_list)}")
    print(f"Energy Alert: {alert_needed}")

# Testing with sample data
school_data = {"Projector": 6, "AC": 9, "Lights": 4}
check_classroom_power(school_data)
```

***Problem 6: Online Event Registration Capacity Controller***

```python

def manage_event_list(max_limit, total_signups):
    # If signups are more than limit, the extra go to waitlist
    if total_signups > max_limit:
        confirmed = max_limit
        waitlist = total_signups - max_limit
        the_status = "Closed"
    else:
        confirmed = total_signups
        waitlist = 0
        the_status = "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlist}")
    print(f"Registration Status: {the_status}")

# Testing with sample data
manage_event_list(100, 105)
```
